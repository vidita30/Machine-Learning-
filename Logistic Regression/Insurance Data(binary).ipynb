```python
#train test split is important to know the score of your model...or how accurate that is 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data[['age']],data.bought_insurance,train_size=0.8)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predicted=model.predict(x_test)
```

```python
#proba is used to find the probability -like for age 45 in x_test probability of not buying,probability of buying has beend shown 
model.predict_proba(x_test)
```

```python
 model.score(x_test,y_test)
```

```python
import numpy as np
plt.xlabel("age")
plt.ylabel("bought_insurance")
plt.scatter(data['age'],data['bought_insurance'],marker='*',color='green')

# Plot logistic regression curve
x_values = np.linspace(min(data['age']), max(data['age']), 100)
y_values = model.predict_proba(x_values.reshape(-1, 1))[:, 1]  # Probability of buying insurance

plt.plot(x_values, y_values, color='blue', label='Logistic Regression Curve')

plt.xlabel('age')
plt.ylabel('Probability of buying insurance')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Generate synthetic classification dataset
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Train logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Scatter plot for data points
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.RdBu, marker='o', s=50, label='Data Points')

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.8, levels=100, antialiased=True)

plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.title("Logistic Regression Decision Boundary on Synthetic Data")
plt.legend()
plt.show()
```
